```python
# prompt: Using dataframe df: Create a distribution using seaborn and the "age" column

import seaborn as sns
sns.displot(df['age'], bins=len(df['age'].unique()))
```

```python
# prompt: Using dataframe df: fill the 'ever_married' column with "Yes" or "No" values. "Yes" should represent 53% of the data and "No" should represent 47% of the data.

import random
df['ever_married'] = ['Yes' if random.random() < 0.53 else 'No' for _ in range(len(df))]
```

```python
# prompt: fill the 'home_state' with the name of a random state found in the United States.

states = ['Alabama', 'Alaska', 'Arizona', 'Arkansas', 'California', 'Colorado', 'Connecticut', 'Delaware',
       'Florida', 'Georgia', 'Hawaii', 'Idaho', 'Illinois', 'Indiana', 'Iowa', 'Kansas', 'Kentucky',
       'Louisiana', 'Maine', 'Maryland', 'Massachusetts', 'Michigan', 'Minnesota', 'Mississippi',
       'Missouri', 'Montana', 'Nebraska', 'Nevada', 'New Hampshire', 'New Jersey', 'New Mexico',
       'New York', 'North Carolina', 'North Dakota', 'Ohio', 'Oklahoma', 'Oregon', 'Pennsylvania',
       'Rhode Island', 'South Carolina', 'South Dakota', 'Tennessee', 'Texas', 'Utah', 'Vermont',
       'Virginia', 'Washington', 'West Virginia', 'Wisconsin', 'Wyoming']

df['home_state'] = [random.choice(states) for _ in range(len(df))]
```

```python
# prompt: Using dataframe df: if the 'age' value is greater than 67 and less than 100, change the 'home_state' value to "Florida"

df.loc[(df['age'] > 67) & (df['age'] < 100), 'home_state'] = 'Florida'
```
